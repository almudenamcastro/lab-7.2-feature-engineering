```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
#your code here
spaceship.shape
```

**Check for data types**

```python
# check for data types
spaceship.info()

# Let's correct the VIP column
spaceship['VIP'] = spaceship['VIP'].astype('bool')

# and define categorial vs numerical columns
cat_var = spaceship.select_dtypes(include="object").columns
num_var = spaceship.select_dtypes(exclude="object").columns

print("Categorical Variables: ", cat_var)
print("Numerical Variables: ", num_var)

len(cat_var) + len(num_var) == len (spaceship.columns)
```

**Check for missing values**

```python
# check for missing values
spaceship.isna().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
# we will drop rows containing any missing value
spaceship = spaceship.dropna(axis=0)
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship['Cabin'].value_counts()
```

```python
# transform cabin in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}
spaceship['Cabin'] = spaceship['Cabin'].str.split('/').str[0]
spaceship['Cabin'].value_counts()
```

- Drop PassengerId and Name

```python
# drop columns PassengerId and Name
spaceship = spaceship.drop(['PassengerId', 'Name'], axis=1)

cat_var = cat_var.drop(['PassengerId', 'Name'])
```

- For non-numerical columns, do dummies.

```python
# do dummies for non-numerical columns 
spaceship = pd.get_dummies(spaceship, columns=cat_var)
```

```python
spaceship.head()
```

**Perform Train Test Split**

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(spaceship.drop('Transported', axis=1), spaceship['Transported'], test_size=0.3, random_state=42)

X_train.head()
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
# We'll use knn classifier (we want to predict the transported column which is categorical)
# Initialize the model

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()

#fit the model
knn.fit(X_train, y_train)
```

- Evaluate your model's performance. Comment it

```python
#evaluate the model 
print("Accuracy:", knn.score(X_test, y_test))
# this is the model accuracy. 


from sklearn.metrics import f1_score

# Predict the labels of the test set
y_pred = knn.predict(X_test)

# Compute the F1 score
f1 = f1_score(y_test, y_pred)

print("F1 score:", f1)

# the accuracy improved thanks to the feature engineering process. Although not much. 
```
